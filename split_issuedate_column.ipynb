```python
import pandas as pd
```

```python
df = pd.read_csv('./foia_data.csv')
df.head()
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
df.head()
```

```python
df['issue_date2'] = df.issue_date.str.split('/')
df.head()
```

```python
df['month'] = df.issue_date2.str.get(0)
df['day'] = df.issue_date2.str.get(1)
df['year'] = df.issue_date2.str.get(2)
df.head()
```

```python
df.to_csv('time_split_ticket_data.csv')
```
